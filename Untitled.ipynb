```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import statistics
```

```python
df = pd.read_csv("supermarket_sales.csv")
df.head(3)
```

```python
df.info() 
```

****Dropping Branch and setting InvoiceID as index column****

```python
df.drop(columns = ["Branch"], inplace = True)
df.head(2)
```

```python
df.set_index("Invoice ID", inplace = True)
```

```python
df.head(3)
```

****Creating a separate column to store the range of ratings****

```python
arr = np.linspace(2, 2+1-0.1, 10)
arr
```

```python
df["Rating_range"] = df["Rating"]
for i in range(1,10):
    arr = np.linspace(i, i+0.9, 10)
    df["Rating_range"].replace(to_replace=arr, value=f"{i}-{i+1}", inplace = True)
```

```python
df.head(2)
```

****Creating columns - Year, Month, Hour, Day, Minute****

```python
df["Datetime"] = pd.to_datetime(df["Date"]+" "+df["Time"])
```

```python
df['Hour'] = df["Datetime"].dt.hour
df['Minute'] = df["Datetime"].dt.minute
df["Year"] = df["Datetime"].dt.year
df["Month"] = df["Datetime"].dt.month
df["Weekday"] = df["Datetime"].dt.day_name()
```

```python
df.head(3)
```

```python
df["Year"].value_counts()
# The data is of 2019 year only
# Therefore we may drop it
```

```python
df.drop(columns = ["Year"], inplace=True)
```

```python
df["Month"].value_counts()   # Only 3 months of data - Jan, Feb, Mar
```

```python
df.info()
```

****Rename Payment column****

```python
df.rename(columns = {"Payment":"Payment_mode"}, inplace = True)
```

****Converting back into csv****

```python
df.to_csv("supermarket_sales_cleaned.csv")
```
